# Darknet

```python
import pandas as pd

# Replace 'your_file.csv' with the path to your actual CSV file
file_path = 'darknet_train.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to ensure it's loaded correctly
# print("First few rows of the DataFrame:")
# print(df.head())

# Count the occurrences of each value in the 'label' column
value_counts = df['label'].value_counts()

# Calculate the percentage for each value
percentages = (value_counts / value_counts.sum()) * 100

# Combine counts and percentages into a single DataFrame
distribution_df = pd.DataFrame({
    'Count': value_counts,
    'Percentage': percentages.round(2)  # Round percentages to 2 decimal places
})

# Display the counts and percentages
print("\nOccurrences and percentages of each value in 'label' column:")
print(distribution_df)

# Display the total count of labels
total_count = value_counts.sum()
print(f"\nTotal count of entries: {total_count}")
```
