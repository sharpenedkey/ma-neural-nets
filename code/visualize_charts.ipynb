```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_paths = {
    'darknet_test': 'darknet_train.csv',
    'kdd_test': 'kdd_train.csv',
    'shuttle_test': 'shuttle_train.csv',
    'covtype_test': 'covtype_train.csv'
}

names = ["Darknet", "KDD", "Shuttle", "Covertype"]

# Create a subplot with 2x2 layout for four pie charts
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

#fig.suptitle("Test Dataset Label Distributions", fontsize=16)

for i, (name, path) in enumerate(file_paths.items()):
    # Load each file
    data = pd.read_csv(path)
    
    # Assuming the column name for labels is 'label' in each file
    label_counts = data['label'].value_counts()
    
    # Get row and column indices for 2x2 layout
    row, col = divmod(i, 2)
    
    # Create pie chart
    wedges, texts = axs[row, col].pie(label_counts, labels=None, startangle=90)
    axs[row, col].set_title(f'{names[i]} Label Distribution', fontsize=16)
    
    # Add legend to the right side of each pie chart
    axs[row, col].legend(wedges, label_counts.index, title="Labels", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Adjust layout
plt.tight_layout()
plt.show()
```
